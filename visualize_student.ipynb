```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv(f'./experiments/base_cnn/train_log.csv')
test_df = pd.read_csv(f'./experiments/base_cnn/test_log.csv')

train_distill_df = pd.read_csv(f'./experiments/cnn_distill/train_log.csv')
test_distill_df = pd.read_csv(f'./experiments/cnn_distill/test_log.csv')
```

```python
x = np.arange(100)

plt.plot(x, train_df['train_acc'], label='Train acc w/o KD')
plt.plot(x, test_df['val_acc'], label='Test acc w/o KD')
plt.plot(x, train_distill_df['train_acc'], label='Train acc w/ KD')
plt.plot(x, test_distill_df['val_acc'], label='Test acc w/ KD')
plt.legend()
plt.show()
```
